```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.corr()
```

```python
df['Ship Mode'].value_counts().plot.bar()
plt.title('GRAFIK BATANG')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.title('GRAFIK BATANG')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.hist(df['Segment'], label ='Segment')
plt.hist(df['Region'], label ='Region')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('GRAFIK PIE')
plt.pie(df['Region'].value_counts(), labels=['West','East','Central','south'], startangle=90, autopct='%.1f%%', explode=(0.1,0.1,0.1,0.1))
plt.legend(loc='lower right')
plt.show()
```

```python
df2 = pd.read_csv('telkom.csv')
df2
```

```python
df2.corr()
```

GRAFIK PLOT & GARIS

Grafik berikut menunjukkan bahwa Saham TLKM naik di bulan ini

```python
plt.figure(figsize=(10,5))
plt.title('GRAFIK PLOT & GARIS')
plt.plot(df2['date'], df2['previous'], label ='Previouse', linestyle ='--')
plt.plot(df2['date'], df2['open_price'], label ='Open Price', linestyle ='--')
plt.plot(df2['date'], df2['high'], label ='High', marker='o')
plt.plot(df2['date'], df2['low'], label ='Low', marker='o')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

```python
plt.boxplot(df2[['high','low']], labels=['rank','puncak'])
plt.title('GRAFIK BOXPLOT')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('SCATTER PLOT')
plt.scatter(df2['date'], df2['first_trade'], label='Sembuh', color = 'coral')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.xticks(rotation=60)
plt.legend()
plt.show()
```

```python
sns.heatmap(data=df2.corr(), cmap ='Reds')
plt.title('GRAFIK HEATMAP')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```
