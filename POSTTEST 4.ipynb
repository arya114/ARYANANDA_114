Lakukan Data cleaning pada data dengan nilai null

```python
df.isna().sum()
```

```python
standarisasi = df.copy()

standarisasi['MarkDown1'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['MarkDown1']])
standarisasi['MarkDown2'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['MarkDown2']])
standarisasi['MarkDown3'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['MarkDown3']])
standarisasi['MarkDown4'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['MarkDown4']])
standarisasi['MarkDown5'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['MarkDown5']])
standarisasi['CPI'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['CPI']])
standarisasi['Unemployment'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['Unemployment']])

standarisasi.isna().sum()
```

Lakukan Data cleaning pada data dengan nilai duplikat
